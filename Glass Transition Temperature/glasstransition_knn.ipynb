```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("glass_data.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_features = scaler.transform(df)
df_scaled=pd.DataFrame(scaled_features,columns = df.columns)
```

```python
X_col=df_scaled.columns.values
```

```python
X=df_scaled[X_col[:-1]]
y=df_scaled['Tg']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25,random_state=100)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import classification_report,confusion_matrix, r2_score, mean_squared_error, make_scorer
from math import sqrt
```

RRMSE: Assesses the predictive performance regression algorithms, bringing the advantage of measuring the
target algorithm improvements over a baseline (a predictor that always predicts the mean value of the desired
output). The RRMSE can be calculated using rrmse function

```python
def rrmse(y_true, y_pred):
    from math import sqrt
    re= sqrt(1-r2_score(y_true, y_pred))
    return re
rrmse_scorer = make_scorer(rrmse,greater_is_better=False)
```

```python
k_limit=12
rrmse_eu=[]
for i in range(1,k_limit):
    eu_knn = KNeighborsRegressor(n_neighbors=i, p=2, metric='minkowski')
    eu_knn.fit(X_train,y_train)
    pred = eu_knn.predict(X_test)
    rrmse_eu.append(rrmse(y_test,pred))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,k_limit),rrmse_eu,color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=5, label="euclidean")
plt.title('RRMSE vs. K Value using KNN-Regressor')
plt.xlabel('K')
plt.ylabel('RRMSE error metric')
```

We see that at k=3, we get the best value.
But this creates problem of leakage from training to test data as we have repeatedly tested it for getting optimal k values so this gets trained on test values as well. To avoid this we have to deploy k-fold cross validation.

```python
from sklearn.model_selection import KFold
```

```python
# Cross_val_score---this is probably faster and can replace traintestsplit if we understand it
# ???
```

```python
# Outer --> K-CV for model evaluation, 
# Inner --> K-CV for hyperparameter tuning

# Define number of folds to test the model=Nouter, Number of folds to tune the hyperparameter=Ninner
Nouter=10 # should be 10
Ninner=5 # should be 5

# outer loop starts here
kf = KFold(n_splits=Nouter)
# Initialte Variable for finding model's predictive performance for outer splits
rrmse_out=[]
rsquare_out=[]
rmse_out=[]
k_used=[]
for train_index, test_index in kf.split(X):
    print(train_index,test_index)
    # Outer Split
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    kf1 = KFold(n_splits=Ninner)
    # inner loop starts here
    # Initiate Variable for hyperparameter and their scoring
    k_opt=[]
    rrmse_opt=[]
    for train_index, test_index in kf1.split(X_train):
        #Inner Split
        X_train_in, X_test_in = X_train.iloc[train_index], X_train.iloc[test_index]
        y_train_in, y_test_in = y_train.iloc[train_index], y_train.iloc[test_index]
        k_limit=12
        rrmse_eu=[]
        for i in range(1,k_limit):
            eu_knn = KNeighborsRegressor(n_neighbors=i, p=2, metric='minkowski')
            eu_knn.fit(X_train_in,y_train_in)
            pred = eu_knn.predict(X_test_in)
            # appends rrmse for each iteration of hyperparameter
            rrmse_eu.append(rrmse(y_test_in,pred))
        # appends best hyperparameter value for given inner split to k_opt 
        k_opt.append(rrmse_eu.index(min(rrmse_eu))+1)
        # appends the lowest rrmse for inner validation set to rrmse_opt
        rrmse_opt.append(min(rrmse_eu))
    
    # finds out the index of best k using lowest rrmse to get the configuration for given outer split
    k_optimal_index=rrmse_opt.index(min(rrmse_opt))
    
    # build the model to test against the best hyperparameter for given outer split
    eu_knn_opt = KNeighborsRegressor(n_neighbors=k_opt[k_optimal_index], p=2, metric='minkowski')
    eu_knn_opt.fit(X_train,y_train)
    pred = eu_knn_opt.predict(X_test)
    
    # Saves model performance metrics and hyperparameter values by taking the best configuration for given outer split
    rrmse_out.append(rrmse(y_test,pred))
    k_used.append(k_opt[k_optimal_index])
    rmse_out.append(mean_squared_error(y_test, pred, squared=False))
    rsquare_out.append(r2_score(y_test, pred))
k_best_value = k_used[rrmse_out.index(min(rrmse_out))]
#this prints out the average and std dev of each metric from 10 folds
print("Best value of k= ",k_best_value,"/n","RRMSE = ",np.mean(rrmse_out),u"\u00B1 ",np.std(rrmse_out),"/n","RMSE = ", np.mean(rmse_out),u"\u00B1 ",np.std(rmse_out),"/n","R\u00b2 ", np.mean(rsquare_out),u"\u00B1 ",np.std(rsquare_out),"/n")
```
